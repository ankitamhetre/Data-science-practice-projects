# Cleaning the data

```python
import numpy as np
from scipy.stats import zscore
z=np.abs(zscore(df))
threshold=3
np.where(z>3)
```

```python
z[25][3]
```

These are the rows*cols with outliers

```python
df_new=df[(z<3).all(axis=1)]

print('original shape',df.shape)
print('new shape',df_new.shape)
```

```python
#percentage loss
#(178-169)/178 * (100)
percentage_loss=(9/1030)*100
print("the percentage loss due to removal of outliers is ",percentage_loss)
```

# removing data with interquartile range IQR

```python
q1=df.quantile(0.25)
q3=df.quantile(0.75)
IQR=q3-q1
print(IQR)
```

```python
df_new1=df[~((df<(q1-1.5*IQR))|(df>(q3+1.5*IQR))).any(axis=1)]
print(df_new1.shape)
```

```python
#percentage loss
#(178-162)/178 * (100)
percentage_loss=((178-162)/178 )*100
print("the percentage loss due to removal of outliers is ",percentage_loss)
```

# Assinging inputs and target variables

```python
df_new.sample()
```

```python
x=df_new.drop('Alcohol',axis=1)
y=df_new['Alcohol']  #target variable
```

# Standard scaling the values

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
scaled_x=sc.fit_transform(x)
```

# MODEL BUILDING

#### Checking for the best Random state 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score

best_rstate=0
accu=0

for i in range(0,200):
    x_train,x_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.25,random_state=i)
    Lr=LinearRegression()
    Lr.fit(x_train,y_train)
    y_pred=Lr.predict(x_test)
    tempaccu=r2_score(y_test,y_pred)
    
    if tempaccu>accu:
        accu=tempaccu
        best_rstate=i
        
        
print('Best accuracy is ',(accu*100),'found on random state',(best_rstate))   
        
        
            
```

#### As the best accuracy is achieved at random state 3 let's use this for our model building

# Linear regression

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.25,random_state=3) #best random state i.e 3 is used
Lr=LinearRegression()
Lr.fit(x_train,y_train)
y_pred=Lr.predict(x_test)
r2score=r2_score(y_test,y_pred)
cvscore=cross_val_score(LinearRegression(),x_train,y_train,cv=5).mean()

print(f"accuracy= {r2score*100} , cross validation = {cvscore} , difference ={(r2score*100)-(cvscore*100)}")
```

# RandomForest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
RF=RandomForestRegressor()
RF.fit(x_train,y_train)
y_pred=RF.predict(x_test)
r2score=r2_score(y_test,y_pred)
cvscore=cross_val_score(RandomForestRegressor(),x_train,y_train,cv=5).mean()

print(f"accuracy= {r2score*100} , cross validation = {cvscore} , difference ={(r2score*100)-(cvscore*100)}")
```

# AdaBoostRegressor

```python
from sklearn.ensemble import AdaBoostRegressor
AD=AdaBoostRegressor()
AD.fit(x_train,y_train)
y_pred=AD.predict(x_test)
r2score=r2_score(y_test,y_pred)
cvscore=cross_val_score(AdaBoostRegressor(),x_train,y_train,cv=5).mean()

print(f"accuracy= {r2score*100} , cross validation = {cvscore} , difference ={(r2score*100)-(cvscore*100)}")
```

# SGDRegressor

```python
from sklearn.linear_model import SGDRegressor
SGD=SGDRegressor()
SGD.fit(x_train,y_train)
y_pred=SGD.predict(x_test)
r2score=r2_score(y_test,y_pred)
cvscore=cross_val_score(SGD,x_train,y_train,cv=5).mean()

print(f"accuracy= {r2score*100} , cross validation = {cvscore} , difference ={(r2score*100)-(cvscore*100)}")
```

# DecisionTree Regression

```python
from sklearn.tree import DecisionTreeRegressor

dtr= DecisionTreeRegressor()
dtr.fit(x_train,y_train)
y_pred=dtr.predict(x_test)
r2score=r2_score(y_test,y_pred)
cvscore=cross_val_score(dtr,x_train,y_train,cv=5).mean()

print(f"accuracy= {r2score*100} , cross validation = {cvscore} , difference ={(r2score*100)-(cvscore*100)}")
```

# GradientBoosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
gb= GradientBoostingRegressor()
gb.fit(x_train,y_train)
y_pred=gb.predict(x_test)
r2score=r2_score(y_test,y_pred)
cvscore=cross_val_score(gb,x_train,y_train,cv=5).mean()

print(f"accuracy= {r2score*100} , cross validation = {cvscore} , difference ={(r2score*100)-(cvscore*100)}")

```

#### From the above observations we can say that the AdaBoost regression works well as the difference between r2score and cross_val is minimum
